여기서는 단순하게 Pillow를 이용해 비교연산으로 중복파일을 제거

```python
unique_list_src = copy.deepcopy(item_list_src)
```

```python
len(unique_list_src)
```

```python
count = 0
for i in range(len(item_list_src)):
    
    item = item_list_src[i]
    if i + 1 == len(item_list_src):
        break
        
    others = item_list_src[i+1:]
    
    for oth in others:
        
        if Image.open(item) == Image.open(oth):
            print(item, " == " ,oth)
            unique_list_src.remove(item)
            print(item, " removed")
            print("-" * 50)
            count += 1
            break
            
        else: 
            continue
            
print("{}/{} items removed".format(count, len(item_list_src)))
```

```python
# 실제 파일 제거 
for item in item_list_src:
    if item not in unique_list_src:
        os.remove(item)
    else:
        continue
```
